```python
#some libraries i always import while scraping
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from bs4 import NavigableString
import json
import time
#libraries i am using for special case 
import pytesseract #to extract data from image
from PIL import Image#to open or read image
```

```python
#function to get text from image
def resolve(path):
    print("resampling image")
    #check_output(['convert',path,'-resample','600',path])
    return pytesseract.image_to_string(Image.open(path))
#initializing dictionary
dict1 = { 0: { },1: { },2: { },3: { }}
#instance to use webdriver with selenium
browser = webdriver.Chrome('/home/rajan/Downloads/chromedriver')
browser.implicitly_wait(30)
#creating list of CIN of companies whose information is needed 
company=['U40108HP1995PTC016183','U40101HP1997PTC020577','U70200TG2009PTC063770','U15332MH1996PTC099964']
     
```

```python
for k in range(len(company)):
    #get webpage
    browser.get('http://www.mca.gov.in/mcafoportal/viewCompanyMasterData.do')
    time.sleep(5)
    #get element consist of captcha image
    source=browser.find_element_by_xpath('//*[@id="captcha"]')
    #location of element which we get above
    loc1=source.location
    #size of element which we get above
    size1=source.size
    #saving webpage screenshot
    browser.save_screenshot("screenshot.png")
    #openning image 
    image=Image.open('screenshot.png')
    #now we will crop image which will consist captcha image
    left=loc1['x']
    top=loc1['y']
    bottom=loc1['y']+size1['height']
    right=loc1['x']+size1['width']
    image2=image.crop((left,top,right,bottom))
    image2.save("/home/rajan/Downloads/image2.png")
    #now put cropped image to function which will extract text from image
    yup=resolve("/home/rajan/Downloads/image2.png")
    print(yup)
    #now submit form 
    #first CIN 
    browser.find_element_by_xpath("//*[@id='companyID']").send_keys(company[k])
    #second extracted text of captcha
    browser.find_element_by_xpath("//*[@id='userEnteredCaptcha']").send_keys(yup)
    time.sleep(10)
    #finally click on submit button to post request and get information page
    browser.find_element_by_xpath("//*[@id='companyLLPMasterData_0']").click()
    #now get elements from which we will get information
    rajan=browser.find_elements_by_xpath("//*[@id='resultTab1']/tbody/tr/td") #master data elements
    keyoid=browser.find_elements_by_xpath("//*[@id='resultTab6']/tbody/tr/th") #table header of Signatory Details table
    valoid=browser.find_elements_by_xpath("//*[@id='resultTab6']/tbody/tr/td") #detail of signatory details table
    d=browser.find_elements_by_xpath("//*[@id='aShowDirectorMasterdata']") 
    box_size=browser.find_elements_by_xpath("//*[@id='resultTab6']/tbody/tr") #number of rows in signatory table
    b=0
    for i in box_size:
        b=b+1
    print(b)
    html_=browser.page_source
    soop = BeautifulSoup(html_, 'html.parser')
    info=soop.find_all('table',{'id':'resultTab6'})
    #initializing some empty lists
    lst=[]
    lst2=[]
    lst3=[]
    for i in valoid:
        #get text from element
        lst3+=[i.text]
    for i in keyoid:
        #get text from element
        lst2+=[i.text]
    for i in rajan:
        #get text from element
        lst+=[i.text]
    print(len(lst))    
    #now we have all data required put them into dictionary in a format
    for i in range(len(lst)):
        if i==0:
            name=lst[i]
            x=i+1
            value=lst[x]
            dict1[k][name]=value
        elif i%2==0:
            name=lst[i]
            x=i+1
            value=lst[x]
            dict1[k][name]=value
        else:
            continue
    h=len(lst3)/len(lst2)
    one=lst2[0]
    two=lst2[1]
    three=lst2[2]
    four=lst2[3]
    five=lst2[4]
    print(len(lst3))
    #initialize collect list to get condition for our structure
    collect=[]
    for i in range(b):
        collect+=[i*len(lst2)]
    print(type(collect[0]))    
    for i in range(0,(len(lst3))):
        if i in collect:
            dict1[k][one+str(i)]=lst3[i]
            dict1[k][two+str(i)]=lst[i+1]
            dict1[k][three+str(i)]=lst3[i+2]
            dict1[k][four+str(i)]=lst3[i+3]
            dict1[k][five+str(i)]=lst3[i+4]
        else:
            continue
print(dict1)       
```
